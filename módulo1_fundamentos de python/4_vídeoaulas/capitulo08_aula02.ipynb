````python
arquivo = open('cidades.txt', 'r')
linhas = arquivo.readlines()
arquivo.close()
print(linhas)
type(linhas)
````

````python
novas_linhas = []
for linha in linhas:
    novas_linhas.append(linha.rstrip())
print(novas_linhas)
````

````python
arquivo = open('cidades.txt', 'r')
for linha in arquivo:
    print(linha.rstrip())
arquivo.close()
````

````python
arquivo = open('cidades.txt', 'r')
soma = 0
for linha in arquivo:
    cidade = linha.split(' ')
    populacao = int(cidade[-1])
    print(populacao)
    soma = soma + populacao
arquivo.close()
print('Total da soma final:', soma)
````
